```python
import os
from agent_framework.azure import AzureOpenAIChatClient

endpoint = "https://azure-ai-foundry-pm01.cognitiveservices.azure.com/"
model_name = "gpt-4.1"
deployment = "gpt-4.1"


api_version = "2024-12-01-preview"

client = AzureOpenAIChatClient(
    api_version=api_version,
    endpoint=endpoint,
    api_key=subscription_key,
    deployment_name=deployment
)
```

```python
agent = client.create_agent(
    instructions="You are a travel planner",
    name="Travel-Agent",
)
```

```python
async def chat_with_agent():
    print("chatting with agent. Type exit to quit")
    while True:
        user_input = input("You: ")
        if user_input == 'exit':
            break
        response = await agent.run(user_input)
        print(f"Agent: {response.text}")

await chat_with_agent()
```

```python
async def streaming_chat():
    async for update in agent.run_stream("Tell me about the travel plan"):
        if update.text:
            print(update.text, end="", flush=True)
        print()
        
await(streaming_chat())
```
